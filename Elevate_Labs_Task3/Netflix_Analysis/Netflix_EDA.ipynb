```python
import pandas as pd     # Defines 'pd'
import seaborn as sns    # Defines 'sns'
import matplotlib.pyplot as plt # Defines 'plt'

# Now load your dataset
df_netflix = pd.read_csv('netflix_titles.csv')

# 2. Analyze Categorical Features (Type: Movie vs TV Show)
sns.countplot(x='type', data=df_netflix, palette='Set2')
plt.title("Distribution of Movies vs TV Shows")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df_netflix['release_year'])
plt.title("Box Plot of Release Years (Finding Outliers)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
# We use .select_dtypes to only get numbers for the correlation matrix
sns.heatmap(df_netflix.select_dtypes(include=['number']).corr(), annot=True, cmap='Reds')
plt.title("Correlation Heatmap")
plt.show()
```
